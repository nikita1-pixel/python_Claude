```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales.csv")
df["region"] = df["region"].fillna("Unknown")
df["salesperson"] = df["salesperson"].fillna("Unknown")
df["product"] = df["product"].str.strip().str.title()
df["revenue"] = df["total"] * (1 - df["returned"])
```

revenue by region

```python
region_rev = df.groupby("region")["revenue"].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
plt.bar(region_rev.index, region_rev.values, color="steelblue")
plt.title("Revenue by Region")
plt.ylabel("Revenue (₹)")
plt.tight_layout()
plt.show()

region_rev
```

salesperson breakdown


```python
sales_rev =df.groupby("salesperson")["revenue"].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
plt.bar(sales_rev.index, sales_rev.values, color="coral")
plt.title("Revenue by Salesperson")
plt.ylabel("Revenue (₹)")
plt.tight_layout()
plt.show()

sales_rev
```

Category breakdown

```python
cat_rev = df.groupby("category")["revenue"].sum()

plt.figure(figsize=(5, 5))
plt.pie(cat_rev.values, labels=cat_rev.index, autopct="%1.1f%%", startangle=90)
plt.title("Revenue by Category")
plt.show()
```

Key Insights
  - **South** is the top region by revenue
  - **Liam** leads all salespersons
  - **20% return rate** — Electronics has the highest returns
  - **June** was the busiest month
